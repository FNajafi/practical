```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.random.rand(100, 2) * 10 
```

```python
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(data)
```

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
# Plot the original data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1])
plt.title('Original Data')
```

```python
# Plot the transformed data after PCA
plt.subplot(1, 2, 2)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.title('Data after PCA')

plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(42)
data = np.random.rand(50, 3) * 10
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c='b', marker='o', label='Original Data')
```

```python
pca = PCA(n_components=3)
pca.fit(data)
```

```python
components = pca.components_
```

```python
ax.quiver(0, 0, 0, components[0, 0], components[0, 1], components[0, 2], color='r', label='PC1')
ax.quiver(0, 0, 0, components[1, 0], components[1, 1], components[1, 2], color='g', label='PC2')
ax.quiver(0, 0, 0, components[2, 0], components[2, 1], components[2, 2], color='y', label='PC3')

ax.set_xlabel('Number of Bedrooms')
ax.set_ylabel('Square Footage')
ax.set_zlabel('Neighborhood Quality')
ax.set_title('PCA: Principal Components')
ax.legend()

plt.show()
```
