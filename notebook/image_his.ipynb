# Gray Scale image Histogram

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image from the "data" directory
image = cv2.imread('../data/gray.jpeg', cv2.IMREAD_GRAYSCALE)

# Calculate the histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Convert the histogram to a 1D numpy array
hist = np.ravel(hist)

# Create a histogram plot
plt.plot(hist, color='black')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Grayscale Image Histogram')

# Display the histogram
plt.show()
```

# Color image histogram

```python

# Load a color image (replace 'your_image.jpg' with the path to your image)
image = cv2.imread('../data/color.jpeg')

# Split the image into its color channels (B, G, R)
b, g, r = cv2.split(image)

# Calculate histograms for each color channel
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

# Plot the histograms
plt.figure(figsize=(10, 6))
plt.subplot(131)
plt.title("Blue Histogram")
plt.plot(hist_b, color='blue')
plt.xlim([0, 256])

plt.subplot(132)
plt.title("Green Histogram")
plt.plot(hist_g, color='green')
plt.xlim([0, 256])

plt.subplot(133)
plt.title("Red Histogram")
plt.plot(hist_r, color='red')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python

# Load a color image (replace 'your_image.jpg' with the path to your image)
image = cv2.imread('../data/color.jpeg')

# Split the image into its color channels (B, G, R)
b, g, r = cv2.split(image)

# Calculate histograms for each color channel
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

# Plot all color histograms in one graph with overlay
plt.plot(hist_b, color='blue', label='Blue')
plt.plot(hist_g, color='green', label='Green')
plt.plot(hist_r, color='red', label='Red')

plt.title("Overlayed Color Histograms")
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.xlim([0, 256])
plt.legend()

plt.show()

```
